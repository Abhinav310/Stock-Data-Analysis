```python
import pandas as pd
import matplotlib.pyplot as mp
from matplotlib import style
style.use('ggplot')
import numpy as np
import seaborn as sns
import pandas_datareader as pdr
import datetime
import matplotlib.dates as mdates

%matplotlib inline
```

```python
mp.figure(1 , figsize = (20,10))

mp.subplot(221)
sns.heatmap(nvda.isnull() , yticklabels = False , cmap = 'viridis' , cbar = False)
mp.title('NVIDIA')

mp.subplot(222)
sns.heatmap(intc.isnull(), yticklabels = False , cmap = 'viridis' , cbar = False)
mp.title('INTEL')

mp.subplot(223)
sns.heatmap(qcom.isnull() , yticklabels = False , cmap = 'viridis' , cbar = False)
mp.title('QUALCOMM')

mp.subplot(224)
sns.heatmap(ibm.isnull() , yticklabels = False , cmap = 'viridis' , cbar = False)
mp.title('IBM')
```

```python
ibm_19 = ibm.loc[pd.Timestamp('2019-01-01'):pd.Timestamp('2019-12-31')]
mp.plot(ibm_19.index , ibm_19['Adj Close'] , color = 'darkorange')
mp.tick_params(labelrotation = 90)
mp.title('IBM STOCK DATA 2019')
```

```python
mp.figure(1 , figsize = (20,10) )

nvda_19 = nvda.loc[pd.Timestamp('2019-1-1'):pd.Timestamp('2020-12-31')]
mp.subplot(221)
mp.plot(nvda_19.index  ,nvda_19['Adj Close'] ,'.'  )
mp.title('NVIDIA STOCK DATA 2019');

intc_19 = intc.loc[pd.Timestamp('2019-1-1'):pd.Timestamp('2020-12-31')]
mp.subplot(222)
mp.plot(intc_19.index , intc_19['Adj Close'] ,'.'  )
mp.title('INTC STOCK DATA 2019');

qcom_19 = qcom.loc[pd.Timestamp('2019-1-1'):pd.Timestamp('2020-12-31')]
mp.subplot(223)
mp.plot(qcom_19.index , qcom_19['Adj Close'] , '.' )
mp.title('QUALCOMM STOCK DATA 2019');

ibm_19 = ibm.loc[pd.Timestamp('2019-1-1'):pd.Timestamp('2020-12-31')]
mp.subplot(224)
mp.plot(ibm_19.index , ibm_19['Adj Close'] ,'.' )
mp.title('IBM STOCK DATA 2019');


```

```python
monthly_nvda_19 = nvda_19.resample('3M').mean()
mp.plot(monthly_nvda_19.index , monthly_nvda_19['Adj Close'] ,'-o', color = 'darkorange')
mp.tick_params(labelrotation = 90);
```

```python
mp.figure(1 , figsize =(20,10))

mp.subplot(221)
mp.plot(monthly_nvda_19.index ,monthly_nvda_19['Adj Close'],'-o')
mp.title('NVIDIA MONTHLY STOCK DATA');

monthly_intc_19 = intc_19.resample('3M').mean()
mp.subplot(222)
mp.plot(monthly_intc_19.index ,monthly_intc_19['Adj Close'],'-o')
mp.title('INTEL MONTHLY STOCK DATA');

monthly_qcom_19 = qcom_19.resample('3M').mean()
mp.subplot(223)
mp.plot(monthly_qcom_19.index ,monthly_qcom_19['Adj Close'],'-o')
mp.title('QUALCOMM MONTHLY STOCK DATA');

monthly_ibm_19 = ibm_19.resample('3M').mean()
mp.subplot(224)
mp.plot(monthly_ibm_19.index ,monthly_ibm_19['Adj Close'],'-o')
mp.title('IBM MONTHLY STOCK DATA');

```
